Within a single figure, create two plots **closely together (next to each o other horizontally)**:
- births and deaths (two lines, based on the columns **Natural increase - Births** and **Natural decrease - Deaths** in the sheet)
- emigration and immigration (two lines, based on the columns **Foreign permanent migration - Immigration** and **Foreign permanent migration - Emigration** in the sheet)

Add legends to each graph explaining the meaning of each line.

```python
import pandas as pd

file_path = 'C:\\Users\\praha\\CodersLab-Course-Python-Data-Analysis\\SESSION 4 preparation exercises\\Webscraping_and_plots_-_Exercise_files\\02_Charts\\01_Data for plots\\Population in Poland - projections.xlsx'

df = pd.read_excel(file_path, sheet_name='natural and migratory movement')

df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,5))

plt.subplot(1, 2, 1) # We set the first plot as the active plot >>> 1 riadok, 2 stlpec, 1.graf
plt.plot(df['Year'], df['Natural increase - Births'], label='Births')
plt.plot(df['Year'], df['Natural decrease - Deaths'], label='Deaths')
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.title('Births and Deaths Over Years')
plt.legend()
 
plt.subplot(1, 2, 2) # We set the second plot as the active plot >>> 1 riadok, 2 stlpec, 2.graf
plt.plot(df['Year'], df['Foreign permanent migration - Immigration'], label='Immigration', color='brown')
plt.plot(df['Year'], df['Foreign permanent migration - Emigration'], label='Emigration', color='green')
plt.xlabel('Year')
plt.ylabel('Number of People')
plt.title('Immigration and Emigration Over Years')
plt.legend()

plt.tight_layout()
 
```
