### Hints

- use your browsers DevTools to find the top panel with names of categories,
- find the panel using its `id`,
- it may be useful to use the `data-depth` attribute to find the link in `a` tag,
- to get the category name you need to remove the string: `'\ue313\n\ue316\n\n\n'` from the text.

````python
# import required libraries here

import requests
from bs4 import BeautifulSoup

website = requests.get('https://prod-kurs.coderslab.pl/index.php')
soup = BeautifulSoup(website.text, 'html.parser')

soup
````

````python
categories = soup.find_all(class_ = 'category')

for i in categories:
    print(i)
````

RIESENIE OD CHATGPT

````python
# Find all 'li' elements with class 'category' and id starting with 'category-'
categories = soup.find_all('li', class_='category', id=lambda x: x and x.startswith('category-'))

# Iterate over each category found
for category in categories:
    # Find the first 'a' element with 'data-depth' attribute equal to '0' within the category
    category_link = category.find('a', {'data-depth': '0'})
    
    # Check if a valid category link is found
    if category_link:
        # Extract and clean up the category name
        category_name = category_link.text.replace('\ue313\n\ue316\n\n\n', '').strip()
        
        # Extract the URL of the category link
        category_url = category_link['href']
        
        # Print the category name and URL
        print(category_name)
        print(category_url)
````

RIESENIE cez top menu od ANET

````python
def get_categories_urls(url):
    
    # Find the panel with category names
    categories_panel = soup.find('ul', id='top-menu')
    url = 'https://prod-kurs.coderslab.pl/index.php'
    
    # Extract category names and URLs
    categories = []
    for link in categories_panel.find_all('a', {"data-depth": "0"}):
        name = link.text.replace('\ue313\n\ue316\n\n\n', '').strip()
        href = link['href']
        categories.append({"url": href, "name": name})

    return categories
````

````python
url = 'https://prod-kurs.coderslab.pl/index.php'
categories_urls = get_categories_urls(url)

print(categories_urls)
````
