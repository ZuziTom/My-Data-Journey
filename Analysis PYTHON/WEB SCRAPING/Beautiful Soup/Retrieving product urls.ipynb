````python
# import required libraries here
import requests
from bs4 import BeautifulSoup

r = requests.get('https://mystore-testlab.coderslab.pl/index.php?id_category=3&controller=category')
soup = BeautifulSoup(r.text, 'html.parser')
````

````python
products = soup.find_all(id= 'js-product-list')

for i in products:
    print(i.prettify())
````

````python
# Find the panel with category names

products = soup.find(id="js-product-list")
categories = products.find_all(class_="js-product product col-xs-6 col-xl-4")

result = []

# Extract URLs
for category in categories:
    a_element = category.find('a')
    if a_element:
        url = a_element['href']
        result.append(url)
print(result)
````
