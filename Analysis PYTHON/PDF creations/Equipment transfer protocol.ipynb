Write a script that will generate a simple document - an equipment transfer protocol. Get the current date automatically, and the rest of the information including the Company Tax Id (NIP) and registration number (REGON) and the employee personal id number (PESEL) from variables defined below. Style the individual elements as you see fit. Save the document as `protocol.pdf`.

Document sample:
```
                                PLACE, DATE
            Employee equipment agreement

Employer:
NAME OF COMPANY
COMPANY ADDRESS
NIP: ...
REGON: ...

Employee:
NAME AND SURNAME
PESEL

The following equipment was handed over on DATE:
- ITEM 1
- ITEM 2
- ITEM 3



................................
Date and employer signature





................................
Date and employee signature
```

````python
CITY = 'Warszawa'
COMPANY_NAME = 'Sample company name'
COMPANY_ADDRESS = 'Company address'
COMPANY_NIP = '9898767654'
COMPANY_REGON = '565434321'

EMPLOYEE_NAME = 'John Connor'
EMPLOYEE_PESEL = '89122145345'

HARDWARE = ['Laptop XYZ ABC with a serial number: RTY7890', 'Keyboard', 'Mouse']
````

````python
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas
import datetime

def generate_protocol(city, company_name, company_address, company_nip, company_regon, 
                      employee_name, employee_pesel, hardware_list):
    # Create a PDF file
    pdf_file = 'protocol.pdf'
    
    pdf = canvas.Canvas(pdf_file, pagesize=letter)

    # Set font size and style
    pdf.setFont("Helvetica", 12)

    # Write "Place, Date" centered at the top
    pdf.setFont("Helvetica-Bold", 12)
    pdf.drawString(250, 720, f"{city}, {datetime.datetime.now().strftime('%Y-%m-%d')}")
    
    # Write "Employee Equipment Agreement" 
    pdf.setFont("Helvetica-Bold", 12)
    pdf.drawString(220, 700, "Employee Equipment Agreement")

    # Write "Employer:" in bold
    pdf.setFont("Helvetica-Bold", 12)
    pdf.drawString(70, 600, "Employer:")
    pdf.setFont("Helvetica", 12)

    # Write the rest of the Employer information in regular font
    pdf.drawString(70, 585, company_name)
    pdf.drawString(70, 570, company_address)
    pdf.drawString(70, 555, f"NIP: {company_nip}")
    pdf.drawString(70, 540, f"REGON: {company_regon}")

    # Write Employee information
    pdf.setFont("Helvetica-Bold", 12)
    pdf.drawString(70, 500, "Employee:")
    pdf.setFont("Helvetica", 12)

    pdf.drawString(70, 485, employee_name)
    pdf.drawString(70, 470, f"PESEL: {employee_pesel}")


     # Write equipment handover details
    pdf.setFont("Helvetica-Bold", 12)
    pdf.drawString(70, 430, f"The following equipment was handed over on {datetime.datetime.now().strftime('%Y-%m-%d')}:")
    pdf.setFont("Helvetica", 12)

    y_position = 415
    for item in hardware_list:
        pdf.drawString(70, y_position, f"- {item}")
        y_position -= 15

   # Draw dots for employer's signature above the label
    pdf.drawString(70, 300, '.' * 80)

    # Draw dots for employee's signature above the label
    pdf.drawString(70, 150, '.' * 80)

    # Write signature labels
    pdf.drawString(70, 290, "Date and employer signature")
    pdf.drawString(70, 140, "Date and employee signature")

    pdf.save()

# Example use
generate_protocol(
    city='Warszawa',
    company_name='Sample company name',
    company_address='Company address',
    company_nip='9898767654',
    company_regon='565434321',
    employee_name='John Connor',
    employee_pesel='89122145345',
    hardware_list=['Laptop XYZ ABC with a serial number: RTY7890', 'Keyboard', 'Mouse']
)
````
