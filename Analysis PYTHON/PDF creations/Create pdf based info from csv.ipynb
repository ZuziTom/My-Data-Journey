Define a function to create PDF for each student

````python
import pandas as pd
from reportlab.lib.pagesizes import letter
from reportlab.pdfgen import canvas

def create_student_page(c, student_data):
    # Format data
    name = student_data['Name']
    surname = student_data['Surname']
    honors = student_data['Honors']
    birth_year = student_data['Birth Year']
    grades = student_data.drop(['Name', 'Surname', 'Honors', 'Birth Year']).to_dict()
    
    # Determine header based on Honors column
    if honors == 'Y':
        header = f"Honors Student - {name} {surname}"
    else:
        header = f"Student - {name} {surname}"
    
    # Set up content
    c.setFont("Helvetica-Bold", 16)
    c.drawString(100, 750, header)
    
    c.setFont("Helvetica", 12)
    c.drawString(100, 730, f"Year of birth: {birth_year}")
    
    c.setFont("Helvetica-Bold", 14)
    c.drawString(100, 700, "Marks:")
    
    c.setFont("Helvetica", 12)
    y_position = 680
    for subject, grade in grades.items():
        c.drawString(120, y_position, f"{subject}: {grade}")
        y_position -= 20

    c.showPage()  # Move to the next page

def main():
    # Read CSV file into a pandas DataFrame
    df = pd.read_csv('marks.csv')
    
    # Create a PDF
    c = canvas.Canvas("data.pdf", pagesize=letter)
    
    # Iterate over each student row
    for index, row in df.iterrows():
        create_student_page(c, row)
    
    c.save()

if __name__ == "__main__":
    main()
````
