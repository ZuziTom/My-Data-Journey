# Highest and lowest price

Using `sort_values` and the data from the _product_prices_renamed.csv_ the file answer the questions below — use data aggregated for entire Poland (`province=='POLAND'`):

1. What product had the highest 'PLN' price in history? (remember to remove the value = 3000)?
1. What product achieved the lowest 'PLN' price in history? (remember to filter out values = 0),

Sort the data according to the following key:

1. by **product_types** column – descending
1. by **value** column – ascending

```python
import pandas as pd
```

```python
# Loading the file
df = pd.read_csv('../../01_Data/product_prices_renamed.csv', sep=';', decimal='.')
df.head()
```

```python
# Filtering data for the entire Poland and in PLN currency, excluding the outlier value 3000 and zero values
df_poland_pln = df[
    (df['province'] == 'POLAND') & 
    (df['currency'] == 'PLN')
]

df_poland_pln_filtered = df_poland_pln[
    (df_poland_pln['value'] != 3000) & 
    (df_poland_pln['value'] != 0)
]
```

```python
# Finding the product with the highest 'PLN' price in history
df_poland_pln_filtered.sort_values(by='value', ascending=False).head(1)
```

```python
# Finding the product with the lowest 'PLN' price in history
df_poland_pln_filtered.sort_values(by='value', ascending=True).head(1)
```

```python
# Sorting the data according to the specified key
df_poland_pln_filtered.sort_values(
    by=['product_types', 'value'], 
    ascending=[False, True]
)
```
