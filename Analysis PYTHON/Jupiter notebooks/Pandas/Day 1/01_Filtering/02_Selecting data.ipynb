> All the product groups needed for this task are in the **product_types** column.

```python
import pandas as pd
```

```python
df = pd.read_csv('../../01_Data/product_prices_renamed.csv',
  sep=';',
  encoding='UTF-8',
  decimal='.'
)

df.head()
```

```python
# What was the average price of a carrot and pea mix per 1 kg in Poland in 2015?

carrot_pea_2015 = df.loc[
    (df['province'] == 'POLAND') &
    (df['product_types'] == 'frozen carrot and pea mix - per 1kg') &
    (df['date'] >= '2015-1') &
    (df['date'] <= '2015-12')

]

print(f"Average Price of Frozen Carrot and Pea Mix per 1 kg in Poland in 2015: {carrot_pea_2015['value'].mean().round(2)} PLN.")

carrot_pea_2015
```

```python
# What was the average price of apple juice in 2016-2018 in the Masovia province?

apple_juice_masovia = df.loc[
    (df['product_types'] == 'apple juice, boxed - per 1l') &
    (df['province'] == 'MASOVIA') &
    (df['date'] >= '2016-1') &
    (df['date'] <= '2018-12')
]

print(f"Average Price of Apple Juice in 2016-2018 in the Masovia Province: {apple_juice_masovia['value'].mean().round(2)}")

apple_juice_masovia
```

```python
%%time
# Average price of tomato paste in Lower Silesia Province in 2003-2015

tomato_paste_lower_silesia = df.loc[
    (df['product_types'] == '30% tomato concentrate - per 1kg') &
    (df['province'] == 'LOWER SILESIA') &
    (df['date'] >= '2003-1') &
    (df['date'] <= '2015-12')
#     (df['date'].dt.year >= 2003) & (df['date'].dt.year <= 2015)
]

print(f"Average Price of Tomato Paste in Lower Silesia Province in 2003-2015: {tomato_paste_lower_silesia['value'].mean().round(2)}")

tomato_paste_lower_silesia
```

```python
%%time
frozen = df.query("product_types == '30% tomato concentrate - per 1kg' and date >= '2003-1' and date <= '2015-12' and province == 'LOWER SILESIA'")

print(f"The average price of tomato paste in Lower Silesia Province in 2003-2015 was {frozen['value'].mean().round(2)} PNL")
```

```python
# Suggest what can be done with values equal to 0. How does this affect the results of point 3?

# To address values equal to 0, which can skew the results, we have several options:

# Exclude Zero Values: 
    # This method involves removing all entries with a value of 0. 
    # This approach assumes that these values are either data entry errors or represent a lack of available price information.
# Impute Zero Values: 
    # This method involves replacing zero values with a reasonable estimate, such as the mean or median price of that product. 
    # This approach can be more reflective of the actual pricing but requires careful consideration to avoid introducing bias.

tomato_paste_lower_silesia = df.loc[
    (df['product_types'] == '30% tomato concentrate - per 1kg') &
    (df['province'] == 'LOWER SILESIA') &
    (df['date'] >= '2003-1') &
    (df['date'] <= '2015-12') &
    (df['value'] != 0)
]

print(f"Average Price of Tomato Paste in Lower Silesia Province in 2003-2015: {tomato_paste_lower_silesia['value'].mean().round(2)}")

tomato_paste_lower_silesia

# Given this result, it's important to examine the dataset for outliers or erroneous entries. 
```
