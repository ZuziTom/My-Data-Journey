````python
import pandas as pd
````

````python
df = pd.read_csv('../../01_Data/product_prices_cleaned.csv', sep=';')
df.head()
````

````python
# Subsection 1: Creating a pivot table with rows as years and columns as product names
pivot_table_lambda = df.pivot_table(index='year', 
                                    columns='product', 
                                    values='value', 
                                    aggfunc='mean',
#                                     aggfunc=lambda x: x.mean()
                                   )

pivot_table_lambda
````

````python
# Subsection 2: Creating a pivot table to see how the mean and median prices were shaped across years
pivot_table_mean_median = df.pivot_table(index='year', 
                                         columns='product', 
                                         values='value', 
                                         aggfunc=['mean', 'median'])

pivot_table_mean_median
````
