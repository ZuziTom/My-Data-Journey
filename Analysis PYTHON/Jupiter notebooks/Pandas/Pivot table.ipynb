````python
import pandas as pd
from pivottablejs import pivot_ui
````

````python

df = pd.read_csv(
                    '../../01_Data/product_prices_cleaned_2.csv', 
                    sep=','
)

df.head(2)
````

````python
pivot_ui(df)
````

````python
pivot_table_mean = df.pivot_table(index='year',
                                  columns='product',
                                  values='value',
                                  aggfunc = 'mean'
)

pivot_table_mean
````

````python
pivot_table_mean = df.pivot_table(index= ['product'],
                                  columns= ['province'],
                                  values= ['value'],
                                  aggfunc = ['mean', 'median']
)

pivot_table_mean
````
