# Highest and lowest price

Using `sort_values` and the data from the _product_prices_renamed.csv_ the file answer the questions below — use data aggregated for entire Poland (`province=='POLAND'`):

1. What product had the highest 'PLN' price in history? (remember to remove the value = 3000)?
1. What product achieved the lowest 'PLN' price in history? (remember to filter out values = 0),

Sort the data according to the following key:

1. by **product_types** column – descending
1. by **value** column – ascending

```python
import pandas as pd
df = pd.read_csv('../../01_Data/product_prices_renamed.csv', sep=';', decimal=',')

df['value'] = pd.to_numeric(df['value'], errors='coerce')

df.head()
```

```python
# Filter data for entire Poland + 3000 & 0 value excluded

df_poland = df.loc[(df['province'] == 'POLAND') & 
                   (df['value'] != 3000) & (df['value'] != 0)]


df_poland
```

```python
df_poland.sort_values(
    by=['product_line', 'value'],
    ascending=[False, True],  # we can define the sorting behavior of each column
    na_position='last'
)
```

```python
# Remove rows where value is 3000 
df_poland.loc[df_poland['value'] != 3000]
```

```python
# FIlter out rows where value is 3000 from DF for entire POLAND province
df_poland = df_poland.loc[df_poland['value'] != 3000]
df_poland
```
