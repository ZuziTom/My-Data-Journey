````python
import pandas as pd
df = pd.read_csv(
                    '../../01_Data/product_prices.csv', 
                    sep=';',
                    decimal=','
)

df.head()
````

````python
df.rename(columns = {'Name': 'province',
 'Product types': 'product_types',
 'Measurement unit': 'currency',
 'Group ID': 'group_id',
 'Goods types': 'product_line',
 'Value': 'value',
 'Date': 'date'})
````
