````python
import json
import requests

APIKEY = '63681360c228597c9d5fe20231521686'
response = requests.get(f'http://api.exchangeratesapi.io/latest?access_key={APIKEY}')
data = response.json()

base_currency = 'EUR'

for currency, rate in data['rates'].items():
    print(f"{rate:.4f} {base_currency}/{currency}")
````

Rozoberme si riadok kódu for currency, rate in data['rates'].items():

data['rates']: data je slovník obsahujúci informácie o výmenných kurzoch.
data['rates'] extrahuje časť slovníka, ktorá sa špecificky týka výmenných kurzov.

.items(): je metóda v Pythone, ktorá vráti zobrazenie s listom dvojíc kľúč-hodnota.
V tomto prípade data['rates'].items() nám dá zobrazenie s dvojicami predstavujúcimi každý kód meny a jeho príslušný výmenný kurz.

for currency, rate in ...:
Toto je for cyklus, ktorý prechádza každou dvojicou kľúč-hodnota v zobrazení získanom z data['rates'].items().
V každej iterácii priradí aktuálny kľúč (kód meny) do premennej currency a aktuálnu hodnotu (výmenný kurz) do premennej rate.

Spájajúc to všetko dohromady, riadok:

for currency, rate in data['rates'].items(): 

v podstate prechádza každou menou a jej príslušným výmenným kurzom v poskytnutom slovníku data. Toto je bežný vzor v Pythone na iteráciu cez dvojice kľúč-hodnota v slovníku alebo položky v iterovateľnom objekte.
