````python
import requests

# Part 1: Ask the user for the target currency and amount to convert

target_currency = input("Enter the currency you want to convert PLN to (e.g., USD, EUR): ")
print(f"Enter the currency you want to convert PLN to (e.g., USD, EUR): {target_currency}" )
amount_pln = float(input("Enter the amount in PLN you would like to convert: "))
print(f"Enter the amount in PLN you would like to convert: {amount_pln} ")


APIKEY = '394e59525c259b029bda192a9af143eb'
URL = f'http://api.exchangeratesapi.io/latest?access_key={APIKEY}'
````

````python
r = requests.get(URL)
data = r.json()
rate = data["rates"]['PLN']
date = data['date']
calculated_amount = target_currency * rate

print(f"{amount_pln} PLN = {rate} {target_currency}")
print(f'Conversion rate is 1 PLN = {rate} {target_currency} as of {date}')
````

````python
try:
    response = requests.get(URL)

    # Part 2: Check if the currency is valid
    if response.status_code == 200:
        data = response.json()
        rate = data['rates'].get(target_currency)

        # If the rate is found, calculate and display the converted amount
        if rate:
            calculated_amount = round(amount_pln * rate, 2)
            print(f"{amount_pln} PLN = {calculated_amount} {target_currency}")
        else:
            print("There is no such currency!")
    else:
        # Inform the user if the entered currency does not exist
        print("There is no such currency!")

except requests.RequestException as e:
    print(f"An error occurred: {e}")
````
