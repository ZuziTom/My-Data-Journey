In the file **wpb_votes_list_2019.csv** you will find data about the votes within the Wroclaw Citizens' Budget, downloaded from the page [Wrocław Open Data](https://www.wroclaw.pl/open-data/dataset/wroclawski-budzet-obywatelski-glosowanie).

Read this data and display a histogram of voters' ages - display bars covering 5-year ranges: voters aged 18 to 23, 24 to 28, etc. (up to 100 years - we assume this is the maximum age of voters).

```python
import pandas as pd

df = pd.read_csv(
                    '../../02_Charts/01_Data for plots/wpb_votes_list_2019.csv', 
                    sep=',',
                    decimal=','
)

df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define age ranges
age_ranges = range(18, 101, 5)

# Create a histogram
plt.hist(df['Age'], bins=age_ranges, edgecolor='black')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Number of Voters')
plt.title("Wroclaw Citizens' Budget Voters' Ages Histogram")

plt.xticks(age_ranges)
plt.grid(axis='y', alpha=0.75)  # horizontalne, pruhlednost

# Show the plot
plt.show()
```
