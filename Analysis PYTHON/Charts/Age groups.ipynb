From the file **Population in Poland - projections.xlsx**, sheet **5-year age groups** read the data for 2021. Display it on a cumulative bar chart.

Add a legend that explains which part of the bars is data on women and which on men.

Under each bar, add a label that tells which age group the bar applies to.

```python
import pandas as pd

file_path = 'C:\\Users\\praha\\CodersLab-Course-Python-Data-Analysis\\SESSION 4 preparation exercises\\Webscraping_and_plots_-_Exercise_files\\02_Charts\\01_Data for plots\\Population in Poland - projections.xlsx'

df = pd.read_excel(file_path, sheet_name='5-year age groups')

df.head()
```

```python
df_2021 = df.query('Year == 2021').reset_index()
df_2021
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))

plt.bar(df_2021['Age'], df_2021['Men'], label='Men')
plt.bar(df_2021['Age'], df_2021['Women'], bottom=df_2021['Men'], label='Women')

plt.ylabel('Population = total')
plt.xlabel('Age group')
 
plt.legend()

# plt.tight_layout()
```
