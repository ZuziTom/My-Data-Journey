Read from the file **Population in Poland - projections.xlsx**, from the sheet **natural and migratory movement** the values of the column 'Foreign permanent migration - Balance'. Display it on a bar chart.

Under each bar add a label with the year.

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'C:\\Users\\praha\\CodersLab-Course-Python-Data-Analysis\\SESSION 4 preparation exercises\\Webscraping_and_plots_-_Exercise_files\\02_Charts\\01_Data for plots\\Population in Poland - projections.xlsx'

df = pd.read_excel(file_path, sheet_name='natural and migratory movement')

df.head()
```

```python


plt.figure(figsize=(15,10))

# Plotting the bar chart
plt.bar(df['Year'], df['Foreign permanent migration - Balance'])

# Adding labels and title
plt.xlabel('Year')
plt.ylabel('Foreign permanent migration - Balance')
plt.title('Foreign Permanent Migration Balance Over Years')

# Set x-axis ticks and labels
plt.xticks(df['Year'], rotation=45)

# Display the plot
plt.show()
```
